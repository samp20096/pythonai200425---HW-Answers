```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
mpg = pd.read_csv('mpg.csv')
mpg
```

# Sorting
---

```python
# Q1.1 -
display(mpg.sort_values(by='mpg', ascending=True))
# Q1.2 - 
display(mpg.sort_values(by='weight', ascending=False))
# Q1.3 -
display(mpg.sort_values(by=['model_year', 'mpg']))
```

# Value Count
---

```python
# Q2.1 - 
display(mpg.value_counts(subset='cylinders'))
# Q2.2 - 
display(mpg.value_counts(subset='origin'))
# Q2.3 - 
display(mpg.value_counts(subset='model_year'))
# model_year 73 has the most cars in it.
```

# GroupBy & Aggregations

```python
# Q3.1 - 
display(mpg.groupby(by='cylinders')[['mpg']].mean().round(2))
# Q3.2 - 
mpg['horsepower'] = pd.to_numeric(mpg['horsepower'], errors='coerce') # fixing the string issue that returns '?'
display(mpg.groupby(by='origin')[['horsepower']].agg(func='max'))
# Q3.3 - 
display(mpg.groupby(by='model_year')[['weight']].agg(func=['mean', 'median']).round(2))
```

# Unique & nunique

```python
# Q4.1 -
display('Total Unique names in data set', mpg['name'].value_counts().count())
# Q4.2 - 
display(mpg['cylinders'].value_counts())
# Q4.3 - 
display(mpg['model_year'].value_counts().unique())
```
